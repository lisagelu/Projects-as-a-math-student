# Aproksimasi Terbaik Pada Ruang Polinomial P[-1,1]

#Define the function domain using NUMPY

```python
import numpy as np
```

```python
y = np.arange(-1,1,0.001)
```

Mendefinisikan polynomial Cheby 

```python
#mengaktifkan polinomial cheby jenis pertama
from sympy import chebyshevt

#n dan x sebagai variabel secara aljabar
from sympy.abc import n,x
```

```python
T4 = chebyshevt(4,x) #polynomial Cheby berderajat n = 4
T11 = chebyshevt(11,x) #polynomial Cheby berderajat n = 11
T18 = chebyshevt(18,x) #polynomial Cheby berderajat n = 18
print('T4 =',T4)
print('T11 =',T11)
print('T18 =',T18)
```

Mentransformsi Polinomial Variabel ke Polinomial yang sifatnya Numerik

```python
#polinomial cheby derajat 4,11,18 pada modul Numpy
TT4 = 8*y**4 - 8*y**2 + 1
TT11 = 1024*y**11 - 2816*y**9 + 2816*y**7 - 1232*y**5 + 220*y**3 - 11*y
TT18 = 131072*y**18 - 589824*y**16 + 1105920*y**14 - 1118208*y**12 + 658944*y**10 - 228096*y**8 + 44352*y**6 - 4320*y**4 + 162*y**2 - 1
print('TT4 =',TT4)
print('TT11 =',TT11)
print('TT18 =',TT18)
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(y,TT4,color='r')
ax.plot(y,TT11,color='g')
ax.plot(y,TT18,color='b')
ax.grid()
```

#Polinomial Monik Cheby derajat 4,11,18

```python
TTT4 = TT4/(2**3)
TTT11 = TT11/(2**10)
TTT18 = TT18/(2**17)
print('TTT4 =',TTT4)
print('TTT11 =',TTT11)
print('TTT18 =',TTT18)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,4))

axes[0].plot(y,TT4,color='r') #Cheby
axes[0].plot(y,TTT4,color='darkgoldenrod') #Monik Cheby
axes[0].set_title("Polinomial Monik Cheby Derajat 4")
axes[0].grid()

axes[1].plot(y,TT11,color='lightgreen') #Cheby
axes[1].plot(y,TTT11,color='olive') #Monik Cheby
axes[1].set_title("Polinomial Monik Cheby Derajat 11")
axes[1].grid()

axes[2].plot(y,TT18,color='lightseagreen') #Cheby
axes[2].plot(y,TTT18,color='b') #Monik Cheby
axes[2].set_title("Polinomial Monik Cheby Derajat 18")
axes[2].grid()
```

#Mendefinisikan sebarang polinomial monik derajat 4,11,18

```python
f4 = y**4 - 0.003*y**3 + 0.00002*y**2 - 1
g4 = y**4 - 0.004*y**2 + 0.00005*y - 1
u4 = y**4 - 0.2*y**3 + 0.000001*y**2 - 1

f11 = y**11 + 0.001*y**9 + 0.00002*y**7 + 0.00000001*y - 1
g11 = y**11 + 0.0022*y**9 + 0.00002*y**6 + 0.00000001*y - 1
u11 = y**11 + 0.001*y**9 + 0.000032*y**4 + 0.00000001*y - 1

f18 = y**18 - 0.00015*y**12 + 0.0000006*y*10 + 0.00000000000000078*y - 1
g18 = y**18 + 0.00000009*y*10 + 0.0000000000000000099*y - 1
u18 = y**18 - 0.0000077*y**12 +0.0000006*y*9+ 0.0000000000000000118*y - 1

print('f4 =',f4)
print('g4 =',g4)
print('u4 =',u4)

print('f11 =',f11)
print('g11 =',g11)
print('u11 =',u11)

print('f18 =',f18)
print('g18=',g18)
print('u18 =',u18)
```

#Monik Cheby vs Monik : Derajat 4

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,4))

axes[0].plot(y,f4,color='lightcoral') #Cheby
axes[0].plot(y,TTT4,color='lightsalmon') #Monik Cheby
axes[0].set_title("Polinomial Monik Cheby vs Monik Derajat 4")
axes[0].grid()

axes[1].plot(y,g4,color='darkslategray') #Cheby
axes[1].plot(y,TTT4,color='cyan') #Monik Cheby
axes[1].set_title("Polinomial Monik Cheby vs Monik Derajat 4")
axes[1].grid()

axes[2].plot(y,u4,color='lime') #Cheby
axes[2].plot(y,TTT4,color='forestgreen') #Monik Cheby
axes[2].set_title("Polinomial Monik Cheby vs Monik Derajat 4")
axes[2].grid()
```
